```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.pngmart.com/sitemap.xml'

response = requests.get(url)
response
```

```python
xml = response.text
xml
```

```python
soup = BeautifulSoup(xml, 'xml')
```

```python
datas = []

for loc in soup.find_all('loc'):
  site_maps = loc.text
  if 'posts' in site_maps:
    print(loc.text)
    datas.append(site_maps)
```

```python
site_maps = datas[0]
```

```python
response = requests.get(site_maps)
response
```

```python
url = response.text
```

```python
soup = BeautifulSoup(url, 'xml')
```

```python
master_list = []
for loc in soup.find_all('loc'):
  files = loc.text
  if 'image' in files:
    print(loc.text)
    master_list.append(files)
```

```python
len(master_list)
```

```python
datas = []
for loc in soup.find_all('loc'):
  url = loc.text
  datas.append(url)
```

```python
master_list[0:5]
```

```python
for image_url in master_list[0:5]:
  print(image_url)
  response = requests.get(image_url)
  soup = BeautifulSoup(response.text, 'html.parser')
  png_url = soup.find('a', {'class':'download'})['href']
  image = requests.get(png_url)
  image_title = image_url.split('/')[-1]+('-')+png_url.split('/')[-1]

  with open('/content/Image_PNG/{}'.format(image_title), 'wb') as file:
    file.write(image.content)
```
