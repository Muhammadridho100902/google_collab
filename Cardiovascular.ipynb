**Penyakit kardiovaskular atau cardiovascular disease (CVD) adalah penyakit yang berkaitan dengan jantung dan pembuluh darah. Penyakit kardiovaskular yang umum adalah penyakit jantung iskemik (ischemic heart disease atau IHD), strok, penyakit jantung akibat tekanan darah tinggi atau penyakit jantung hipertensi (hypertensive heart disease), penyakit jantung reumatik (rheumatic heart disease atau RHD), aneurisma aorta (aortic aneurysm), kardiomiopati (cardiomyopathy), fibrilasi atrium (atrial fibrillation), penyakit jantung bawaan (congenital heart disease), endokarditis, dan penyakit arteri perifer (peripheral artery disease atau PAD). (Sumber: Wikipedia)**

**Faktor-faktor yang mempengaruhi kardiovaskuler meliputi usia, jenis kelamin, hipertensi, diabetes melitus, dislipidemia, obesitas, merokok dan riwayat keluarga123. Semua faktor ini dapat meningkatkan risiko terjadinya penyakit kardiovaskuler.**

```python
from scipy.stats import stats

columns = ['Height_(cm)', 'Weight_(kg)', 'BMI',
       'Fruit_Consumption', 'Green_Vegetables_Consumption',
       'FriedPotato_Consumption']

Q1 = data_dummy[columns].quantile(q=.10)
Q3 = data_dummy[columns].quantile(q=.90)
IQR = data_dummy[columns].apply(stats.iqr)

data_dummy[columns] = data_dummy[columns][~((data_dummy[columns] < (Q1-3.5*IQR)) | (data_dummy[columns] > (Q3+3.5*IQR))).any(axis=1)]
```
